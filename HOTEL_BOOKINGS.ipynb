## Problem Statement

The hospitality industry is facing a significant challenge in accurately predicting hotel booking demand, leading to inefficiencies in pricing and room allocation strategies. This machine learning project aims to address this problem by developing a predictive model to forecast hotel booking demand based on historical data and multiple factors including seasonality, location, price, and amenities. The model will analyze patterns and relationships between these variables to determine the likelihood of a hotel room being booked. The objective is to provide valuable insights for hotel management to optimize pricing and allocation strategies, ultimately increasing revenue and occupancy rates. The project seeks to demonstrate the power of machine learning in enhancing decision making in the hospitality industry

## Data Definition

### Independent Variable

1. ADR: Average Daily Rate.
2. Adults: Number of adults.
3. Agent: ID of the travel agency that made the booking.
4. Arrival_Date_Day_Of_Month: Day of the month of the arrival date.
5. Arrival_Date_Month: Month of arrival date with 12 categories: “January” to “December”.
6. Arrival_Date_Week_Number: Week number of the arrival date.
8. Arriva_lDate_Year: Year of arrival date.
9. Assigned_Room_Type: Code for the type of room assigned to the booking. Sometimes the assigned room type differs from the reserved room type due to hotel operation reasons (e.g. overbooking) or by customer request. Code is presented instead of designation for anonymity reasons.
10. Babies: Number of babies
11. Booking_Changes: Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation.
12. Children: Number of children.
13. Company: ID of the company/entity that made the booking or responsible for paying the booking. ID is presented instead of designation for anonymity reasons.
14. Country:  Country of origin. Categories are represented in the ISO 3155–3:2013 format .
15. Customer_Type: Type of booking, assuming one of four categories:	Contract - when the booking has an allotment or other type of contract associated to it; Group – when the booking is associated to a group; Transient – when the booking is not part of a group or contract, and is not associated to other transient booking; Transient-party – when the booking is transient, but is associated to at least other transient booking.
16. Days_In_Waiting_List: Number of days the booking was in the waiting list before it was confirmed to the customer.
17. Deposit_Type: Indication on if the customer made a deposit to guarantee the booking. This variable can assume three categories. No Deposit – no deposit was made; Non Refund – a deposit was made in the value of the total stay cost; Refundable – a deposit was made with a value under the total cost of stay.
18. Hotel: Type of Hotel booked.
19. Distribution_Channel: Booking distribution channel. The term “TA” means “Travel Agents” and “TO” means “Tour Operators”.
20. Is_Repeated_Guest: Value indicating if the booking name was from a repeated guest (1) or not (0).
21. Lead_Time: Number of days that elapsed between the entering date of the booking into the PMS and the arrival date.
22. Market_Segment: Market segment designation. In categories, the term “TA” means “Travel Agents” and “TO” means “Tour Operators”.
23. Meal: Type of meal booked. Categories are presented in standard hospitality meal packages:Undefined/SC – no meal package; BB– Bed & Breakfast; HB – Half board (breakfast and one other meal – usually dinner);FB – Full board (breakfast, lunch and dinner).
24. Previous_Bookings_Not_Canceled: Number of previous bookings not cancelled by the customer prior to the current booking.
25. Previous_Cancellations: Number of previous bookings that were cancelled by the customer prior to the current booking.
26. Required_Car_Parking_Spaces: Number of car parking spaces required by the customer.
27. Reservation_Status:  Reservation last status, assuming one of three categories:	Canceled – booking was canceled by the customer;Check-Out – customer has checked in but already departed;No-Show – customer did not check-in and did inform the hotel of the reason why.
28. Reservation_Status_Date: Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel.
29. Reserved_Room_Type: Code of room type reserved. Code is presented instead of designation for anonymity reasons.
30. Stays_In_Weekend_Nights: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel.
31. Stays_In_Week-Nights: Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel.
32. Total_Of_Special_Requests: Number of special requests made by the customer (e.g. twin bed or high floor).

### Dependent Variable


Is_Canceled: Value indicating if the booking was canceled (1) or not (0)

```python
# with the following function we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature

def correlation(data, threshold):
    col_corr = set()  # Set of all the names of correlated columns
    corr_matrix = data.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: # we are interested in absolute coeff value
                colname = corr_matrix.columns[i]  # getting the name of column
                col_corr.add(colname)
    return col_corr
```

```python
from sklearn.feature_selection import chi2

x1 = data[['hotel','meal','country', 'market_segment', 'reserved_room_type', 'deposit_type', 'customer_type']]
y1 = data['is_canceled']
```

```python
f_score =chi2(x1,y1)
p_value = pd.Series(f_score[1],index = x1.columns)
p_value = p_value.sort_values(ascending = True)
p_value
```

```python
cols = []

for i in p_value.index:
    if p_value[i] <=0.05:
        print(i, '------------', 'Reject Null Hypothesis')
    else:
        print(i, '------------', 'Accept Null Hypothesis')
        cols.append(i)
```

```python
# droping features that will not add any value to my model 

data.drop(labels = cols, axis = 1, inplace = True)
```

```python
data.drop('reservation_status_date', axis =1, inplace =True)
```

```python
data['Total_guests']=data['adults']+data['children']+data['babies']
```
